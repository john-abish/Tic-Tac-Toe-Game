```python
##### Create Board  #####
board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']  

from IPython.display import clear_output

def display_board(board):
    print(board[7] + '  | ' + board[8] + '  | ' + board[9])
    print('------------')
    print(board[4] + '  | ' + board[5] + '  | ' + board[6])
    print('------------')
    print(board[1] + '  | ' + board[2] + '  | ' + board[3]) 

    

##### Define Player Input  #####
def player_input():
    Mk1=''
    Mk2=''
    
    while Mk1!='x' and Mk1!='o':
        Mk1=input("Player 1 choose 'x' or 'o': ")
    if Mk1=='x':
        Mk2='o'
    else:
        Mk2='x'
    
    clear_output()
    print(f' \n Player 1 is {Mk1}\n Player 2 is {Mk2}')
   
    marker=(Mk1,Mk2)
    return marker


##### Define Winning condition  #####
def win_check(board, marker,i):
    k=1
    for _ in range(3):
        if board[k]==board[k+1]==board[k+2]==marker[0]:
            print('Player 1 has WON the game')
            return True
        elif board[k]==board[k+1]==board[k+2]==marker[1]:
            print('Player 2 has WON the game')
            return True
        k+=3
    
    m=1
    for _ in range(3):
        if board[m]==board[m+3]==board[m+6]==marker[0]:
            print('Player 1 has WON the game')
            return True
        elif board[m]==board[m+3]==board[m+6]==marker[1]:
            print('Player 2 has WON the game')  
            return True
        m+=1

    if board[1]==board[5]==board[9]==marker[0] or board[3]==board[5]==board[7]==marker[0]:
        print('Player 1 has WON the game')
        return True
    elif board[1]==board[5]==board[9]==marker[1] or board[3]==board[5]==board[7]==marker[1]:
        print('Player 2 has WON the game')
        return True
    
    elif i>=8:
        print('The game is a TIE')
    #space=space_check(board)
    #if space!=True:
    #    print('The game is a TIE')
    #    return True


##### Get player input #####
def place_marker_final(board, marker):
    #board=[' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    i=1
    #num1=1
    #num2=1
    display_board(board)
    while i<=9:
        
        while i%2!=0: 
            num1=int(input('Enter play position, Player 1: '))
            while board[num1]==marker[0] or board[num1]==marker[1]:
                num1=int(input('Enter NEW play position, Player 1: '))
            board[num1]=marker[0]
            clear_output()
            display_board(board)
            win_check(board, marker,i)
            break
        
        while i%2==0: 
            num2=int(input('Enter play position, Player 2: '))
            while board[num2]==marker[0] or board[num2]==marker[1]:
                num2=int(input('Enter NEW play position, Player 2: '))
            board[num2]=marker[1]
            clear_output()
            display_board(board)
            win_check(board, marker,i)
            break
        #print(i)
        #print(num1)
        #print(num2)
        i+=1
        #display_board(board)
    clear_output()
    display_board(board)

##### Play Game #####
def play_game():
    #print('Welcome to TIC TAC TOE')
    board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    from IPython.display import clear_output
    #display_board(board)
    marker=player_input()
    place_marker_final(board, marker)
```
